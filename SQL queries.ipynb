```python
from sqlalchemy import text
tables_sql = """
CREATE TABLE IF NOT EXISTS Sport (
    SportID SERIAL PRIMARY KEY,
    SportName VARCHAR(50) NOT NULL
);

CREATE TABLE IF NOT EXISTS Tournament (
    TournamentID SERIAL PRIMARY KEY,
    SportID INT REFERENCES Sport(SportID),
    Name VARCHAR(100) NOT NULL,
    StartDate DATE NOT NULL,
    EndDate DATE NOT NULL,
    MaxTeams INT NOT NULL
);

CREATE TABLE IF NOT EXISTS Team (
    TeamID SERIAL PRIMARY KEY,
    TeamName VARCHAR(100),
    Department VARCHAR(50),
    Coach VARCHAR(100)
);

CREATE TABLE IF NOT EXISTS Player (
    PlayerID SERIAL PRIMARY KEY,
    TeamID INT REFERENCES Team(TeamID),
    PlayerName VARCHAR(100),
    Department VARCHAR(50),
    YearLevel VARCHAR(10)
);

CREATE TABLE IF NOT EXISTS Venue (
    VenueID SERIAL PRIMARY KEY,
    VenueName VARCHAR(100),
    Location VARCHAR(100)
);

CREATE TABLE IF NOT EXISTS Referee (
    RefereeID SERIAL PRIMARY KEY,
    RefereeName VARCHAR(100),
    Qualification VARCHAR(100)
);

CREATE TABLE IF NOT EXISTS `Match` (
    MatchID SERIAL PRIMARY KEY,
    TournamentID INT REFERENCES Tournament(TournamentID),
    Team1ID INT REFERENCES Team(TeamID),
    Team2ID INT REFERENCES Team(TeamID),
    VenueID INT REFERENCES Venue(VenueID),
    RefereeID INT REFERENCES Referee(RefereeID),
    MatchDate TIMESTAMP,
    Team1Score INT,
    Team2Score INT,
    WinnerTeamID INT REFERENCES Team(TeamID)
);

# CREATE TABLE IF NOT EXISTS TeamTournament (
#     TournamentID INT REFERENCES Tournament(TournamentID),
#     TeamID INT REFERENCES Team(TeamID),
#     PRIMARY KEY (TournamentID, TeamID)
# );
"""
with engine.begin() as conn:
    for statement in tables_sql.strip().split(';'):
        stmt = statement.strip()
        if stmt:             
            conn.execute(text(stmt))
print("All tables created!")
```

```python
# Create tournament
def create_tournament():
    print("Create New Tournament")
    sport_id = int(input("Sport ID: "))
    name = input("Name: ")
    start_date = input("Start Date (YYYY-MM-DD): ")
    end_date = input("End Date (YYYY-MM-DD): ")
    max_teams = int(input("Max Teams: "))
    query = """
    INSERT INTO Tournament (SportID, Name, StartDate, EndDate, MaxTeams)
    VALUES (:sport_id, :name, :start_date, :end_date, :max_teams)
    """
    with engine.begin() as conn:
        conn.execute(
            text(query),
            {
                "sport_id": sport_id, "name": name, "start_date": start_date, "end_date": end_date, "max_teams": max_teams
            }
        )
    print("TOURNAMENT CREATED!")


# Edit tournament
def edit_tournament():
    print("Edit Tournament")
    tid = int(input("TournamentID: "))
    new_name = input("New Name: ")
    new_start_date = input("New Start Date (YYYY-MM-DD): ")
    new_end_date = input("New End Date (YYYY-MM-DD): ")
    new_max_teams = input("New Max Teams: ")

    updates = {}
    if new_name:
        updates["Name"] = new_name
    if new_start_date:
        updates["StartDate"] = new_start_date
    if new_end_date:
        updates["EndDate"] = new_end_date
    if new_max_teams:
        updates["MaxTeams"] = int(new_max_teams)

    if not updates:
        print("No attributes to update.")
        return

    set_clause = ", ".join(f"{k} = :{k}" for k in updates)
    updates["TournamentID"] = tid

    query = f"""
    UPDATE Tournament  SET {set_clause}  WHERE TournamentID = :TournamentID
    """

    with engine.begin() as conn:
        conn.execute(text(query), updates)

    print("TOURNAMENT UPDATED!")


#Remove tournament
def remove_tournament():
    print("Remove Tournament")
    tournament_id = int(input("Tournament ID: "))
    with engine.begin() as conn:
        conn.execute(
            text("DELETE FROM `Match` WHERE TournamentID = :tid"),
            {"tid": tournament_id}
        )
        conn.execute(
            text("DELETE FROM TeamTournament WHERE TournamentID = :tid"),
            {"tid": tournament_id}
        )
        conn.execute(
            text("DELETE FROM Tournament WHERE TournamentID = :tid"),
            {"tid": tournament_id}
        )

    print("TOURNAMENT DELETED!")


#List tournaments

def list_tournaments():
    query = "SELECT * FROM Tournament"
    with engine.begin() as conn:
        result = conn.execute(text(query))
        rows = result.fetchall()

    for r in rows:
        print(r)

#Schedule match
from sqlalchemy import text

# ORGANIZER: MANAGE SCHEDULE, VENUE, DATE, REFEREE
def schedule_match():
    print("Schedule Match")
    tournament_id = int(input("Tournament ID: "))
    team1_id = int(input("Team 1 ID: "))
    team2_id = int(input("Team 2 ID: "))
    venue_id = int(input("Venue ID: "))
    referee_id = int(input("Referee ID: "))
    match_date = input("Match Date (YYYY-MM-DD): ")

    query = """
    INSERT INTO `Match` (TournamentID, Team1ID, Team2ID, VenueID, RefereeID, MatchDate)
    VALUES (:tournament_id, :team1_id, :team2_id, :venue_id, :referee_id, :match_date)
    """

    with engine.begin() as conn:
        conn.execute(
            text(query),
            {
                "tournament_id": tournament_id,
                "team1_id": team1_id,
                "team2_id": team2_id,
                "venue_id": venue_id,
                "referee_id": referee_id,
                "match_date": match_date,
            },
        )

    print("Match scheduled!")

def view_schedule_for_tournament():
    tid = int(input("Tournament ID: "))

    query = """
    SELECT
      m.MatchDate,
      t1.TeamName AS Team1,
      t2.TeamName AS Team2,
      v.VenueName,
      r.RefereeName,
      m.Team1Score,
      m.Team2Score,
      CASE
        WHEN m.Team1Score > m.Team2Score THEN t1.TeamName
        WHEN m.Team2Score > m.Team1Score THEN t2.TeamName
        ELSE 'Draw'
      END AS Winner
    FROM `Match` m
    JOIN Team t1 ON t1.TeamID = m.Team1ID
    JOIN Team t2 ON t2.TeamID = m.Team2ID
    JOIN Venue v ON v.VenueID = m.VenueID
    JOIN Referee r ON r.RefereeID = m.RefereeID
    WHERE m.TournamentID = :tid
    ORDER BY m.MatchDate;
    """

    with engine.begin() as conn:
        result = conn.execute(text(query), {"tid": tid})
        rows = result.fetchall()

    for r in rows:
        print(r)


def update_match_results():
    match_id = int(input("Match ID: "))
    team1_score = int(input("Team 1 Score: "))
    team2_score = int(input("Team 2 Score: "))

    if team1_score > team2_score:
        winner_team_id = int(input("Winner Team ID: "))
    elif team2_score > team1_score:
        winner_team_id = int(input("Winner Team ID: "))
    else:
        winner_team_id = None  # draw

    query = """
    UPDATE `Match`
    SET Team1Score = :team1_score,
        Team2Score = :team2_score,
        WinnerTeamID = :winner_team_id
    WHERE MatchID = :match_id
    """

    with engine.begin() as conn:
        conn.execute(
            text(query),
            {
                "team1_score": team1_score,
                "team2_score": team2_score,
                "winner_team_id": winner_team_id,
                "match_id": match_id,
            },
        )

    print("Results updated!")

def department_participation():
    query = """
    SELECT t.Department, COUNT(*) AS Total_Teams
    FROM Team t
    GROUP BY t.Department
    ORDER BY Total_Teams DESC;
    """

    with engine.begin() as conn:
        result = conn.execute(text(query))
        rows = result.fetchall()

    for r in rows:
        print(r)


def sport_participation():
    query = """
    SELECT s.SportName, COUNT(t.TournamentID) AS Total_Tournaments
    FROM Sport s
    LEFT JOIN Tournament t ON s.SportID = t.SportID
    GROUP BY s.SportName
    ORDER BY Total_Tournaments DESC;
    """

    with engine.begin() as conn:
        result = conn.execute(text(query))
        rows = result.fetchall()

    for r in rows:
        print(r)


```

```python
def players_of_team():
    team_id = input("TEAM_ID: ")

   
    query = """
    SELECT p.PlayerName, p.Department, p.YearLevel
    FROM Player p
    JOIN Team t ON t.TeamID = p.TeamID
    WHERE t.TeamID = :team_id;
    """

    with engine.begin() as conn:
        result = conn.execute(text(query), {"team_id": team_id})
        rows = result.fetchall()

    for r in rows:
        print(r)

def tournaments_of_team():
    team_id = input("TEAM_ID: ")
    query = """
    SELECT tr.Name, tr.StartDate, tr.EndDate
    FROM Tournament tr
    JOIN TeamTournament tt ON tt.TournamentID = tr.TournamentID
    WHERE tt.TeamID = :team_id;
    """
    with engine.begin() as conn:
        result = conn.execute(text(query), {"team_id": team_id})
        rows = result.fetchall()

    for r in rows:
        print(r)

def matches_of_team():

    team_id = input("TEAM_ID: ")
    query = """
    SELECT
        m.MatchDate,
        t2.TeamName AS Opponent,
        v.VenueName
    FROM `Match` m
    JOIN Team t1 ON t1.TeamID = m.Team1ID OR t1.TeamID = m.Team2ID
    JOIN Team t2
        ON t2.TeamID IN (m.Team1ID, m.Team2ID)
        AND t2.TeamID <> t1.TeamID
    JOIN Venue v ON v.VenueID = m.VenueID
    WHERE t1.TeamID = :team_id
    ORDER BY m.MatchDate;
    """
    with engine.begin() as conn:
        result = conn.execute(text(query), {"team_id": team_id})
        rows = result.fetchall()

    for r in rows:
        print(r)

#Teams and players can view their performance history by tournament
from sqlalchemy import text

# Teams and players can view their performance history by team
def performance_history():
    team_id = input("TEAM_ID: ")
    query = """
    SELECT
        t.TeamID,
        t.TeamName,
        SUM(
            CASE
                WHEN m.WinnerTeamID = t.TeamID THEN 1
                ELSE 0
            END
        ) AS Wins,
        SUM(
            CASE
                WHEN (m.Team1ID = t.TeamID OR m.Team2ID = t.TeamID)
                     AND m.WinnerTeamID <> t.TeamID
                     AND m.WinnerTeamID IS NOT NULL
                THEN 1
                ELSE 0
            END
        ) AS Losses
    FROM Team t
    LEFT JOIN `Match` m
        ON (m.Team1ID = t.TeamID OR m.Team2ID = t.TeamID)
    WHERE t.TeamID = :team_id
    GROUP BY t.TeamID, t.TeamName;
    """

    with engine.begin() as conn:
        result = conn.execute(text(query), {"team_id": team_id})
        row = result.fetchone()

        print(f"Team: {row[1]}; Wins {row[2]} times; Losses {row[3]} times.")


def roster_count():
    team_id = input("TEAM_ID: ")
    
    query = """
    SELECT
        t.TeamName,
        COUNT(p.PlayerID) AS RosterSize
    FROM Team t
    LEFT JOIN Player p ON p.TeamID = t.TeamID
    WHERE t.TeamID = :team_id
    GROUP BY t.TeamName;
    """
    with engine.begin() as conn:
        result = conn.execute(text(query), {"team_id": team_id})
        row = result.fetchone()
        print(f"Team: {row[0]}, Players: {row[1]}")
    


#Teams can register for tournaments up to the registration deadline
from datetime import date, datetime

def register_team_for_tournament():
    team_id = input("TEAM_ID: ")
    tournament_id = input("TOURNAMENT_ID: ")
    select_sql = """
    SELECT t.StartDate,
           t.MaxTeams,
           (
             SELECT COUNT(*)
             FROM TeamTournament tt
             WHERE tt.TournamentID = t.TournamentID
           ) AS RegisteredTeams
    FROM Tournament t
    WHERE t.TournamentID = :tournament_id
    """
    with engine.begin() as conn:
        result = conn.execute(text(select_sql), {"tournament_id": tournament_id})
        row = result.fetchone()

        start_date, max_teams, registered = row
        if isinstance(start_date, datetime):
            start_date = start_date.date()
        if date.today() > start_date:
            print("Registration closed.")
            return
        if registered >= max_teams:
            print("Tournament is full.")
            return
        insert_sql = """
        INSERT INTO TeamTournament (TeamID, TournamentID)
        VALUES (:team_id, :tournament_id)
        """
        conn.execute(
            text(insert_sql),
            {"team_id": team_id, "tournament_id": tournament_id}
        )

    print("Team registered for tournament!")


def add_player_to_roster():
    team_id = input("TEAM_ID: ")
    tournament_id = input("TOURNAMENT_ID: ")
    select_sql = """
    SELECT t.StartDate
    FROM Tournament t
    JOIN TeamTournament tt ON t.TournamentID = tt.TournamentID
    WHERE tt.TeamID = :team_id AND tt.TournamentID = :tournament_id
    """
    with engine.begin() as conn:
        result = conn.execute(
            text(select_sql),
            {"team_id": team_id, "tournament_id": tournament_id},
        )
        row = result.fetchone()
        start_date = row[0]
        if isinstance(start_date, datetime):
            start_date = start_date.date()
        if date.today() >= start_date:
            print("tournament has started.")
            return

        player_name = input("PLAYER_NAME: ")
        dept = input("DEPARTMENT: ")
        year = input("YEAR: ") 

        insert_sql = """
        INSERT INTO Player (PlayerName, Department, YearLevel, TeamID)
        VALUES (:player_name, :dept, :year, :team_id)
        """

        conn.execute(
            text(insert_sql),
            {
                "player_name": player_name,
                "dept": dept,
                "year": year,
                "team_id": team_id,
            },
        )

    print("Player added!")
    
def remove_player_from_roster():
    team_id = input("TEAM_ID: ")
    player_id = input("PLAYER_ID: ")
    tournament_id = input("TOURNAMENT_ID: ")
    select_sql = """
    SELECT t.StartDate
    FROM Tournament t
    JOIN TeamTournament tt ON t.TournamentID = tt.TournamentID
    WHERE tt.TeamID = :team_id AND tt.TournamentID = :tournament_id
    """
    delete_sql = """
    DELETE FROM Player
    WHERE PlayerID = :player_id AND TeamID = :team_id
    """

    with engine.begin() as conn:
        result = conn.execute(
            text(select_sql),
            {"team_id": team_id, "tournament_id": tournament_id},
        )
        row = result.fetchone()

        if not row:
            print("The team is not registered for that tournament.")
            return

        start_date = row[0]
        if isinstance(start_date, datetime):
            start_date = start_date.date()

        if date.today() >= start_date:
            print("Tournament has started.")
            return

        res = conn.execute(
            text(delete_sql),
            {"player_id": player_id, "team_id": team_id},
        )


    print("Player removed!")


def team():
    while True:
        print("User Team, Please select 1 option: ")
        print("1. View players")
        print("2. View tournament")
        print("3. View matches")
        print("4. View performance history")
        print("5. Roster Count")
        print("6. Register team for tournament")
        print("7. Add players to roster")
        print("8. Remove player from roster")
        print("0. Exit")
        print("===================================")

        choice = input("Your option: ")

        if choice == "1":
            players_of_team()
            break
        elif choice == "2":
            tournaments_of_team()
            break
        elif choice == "3":
            matches_of_team()
            break
        elif choice == "4":
            performance_history()
            break
        elif choice == "5":
            roster_count()
            break
        elif choice == "6":
            register_team_for_tournament()
            break
        elif choice == "7":
            add_player_to_roster()
            break
        elif choice == "8":
            remove_player_from_roster()
            break

        elif choice == "0":
            print("Exiting Organizer Menu...");
            break
        else:
            print("Again.")


team()
```

```python
# Spectators can view match results and upcoming matches

def match_results():
    query = """
    SELECT
        T.Name,
        m.MatchDate,
        t1.TeamName AS Team1,
        t2.TeamName AS Team2,
        m.Team1Score,
        m.Team2Score,
        w.TeamName AS Winner
    FROM `Match` m
    JOIN Tournament T ON m.TournamentID =  T.TournamentID
    JOIN Team t1 ON t1.TeamID = m.Team1ID
    JOIN Team t2 ON t2.TeamID = m.Team2ID
    LEFT JOIN Team w ON w.TeamID = m.WinnerTeamID
    WHERE m.Team1Score IS NOT NULL
      AND m.Team2Score IS NOT NULL
    ORDER BY m.MatchDate DESC;
    """

    with engine.begin() as conn:
        result = conn.execute(text(query))
        rows = result.fetchall()
    print("======MATCH RESULTS======")

    for r in rows:
        print(
            f" {r[0]} | "
            f"{r[1]} {r[3]} - {r[2]} {r[4]} | "
        )

def upcoming_matches():
    query = """
    SELECT
        m.MatchDate,
        t1.TeamName,
        t2.TeamName,
        v.VenueName
    FROM `Match` m
    JOIN Team t1 ON t1.TeamID = m.Team1ID
    JOIN Team t2 ON t2.TeamID = m.Team2ID
    JOIN Venue v ON v.VenueID = m.VenueID
    WHERE m.MatchDate > CURRENT_DATE
    ORDER BY m.MatchDate;
    """
    with engine.begin() as conn:
        result = conn.execute(text(query))
        rows = result.fetchall()
    print("======UPCOMING MATCHES======")
    if not rows:
        print('There are no upcoming matches!')
    for r in rows:
        print(r)

def spectator():
    while True:
        print("User Spectator, Please select 1 option: ")
        print("1. View match results")
        print("2. View upcoming matches")
        print("0. Exit")

       

        choice = input("Your option: ")

        if choice == "1":
            match_results()
            break
        elif choice == "2":
            upcoming_matches()
            break
        elif choice == "0":
            print("Exiting Organizer Menu...");
            break
        else:
            print("Again.")


spectator()
```
